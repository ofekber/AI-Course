```python
import pandas as pd
from IPython.display import display

# Define file paths
files = [
    "My_Results.txt",
    "Ofek_Results.txt",
    "Sonya_Results.txt",
    "Lior_Results.txt"
]

# Load and parse data
dataframes = []
for file in files:
    with open(file, "r") as f:
        lines = f.readlines()
    # Extract rounds, accounting for variations in format
    results = []
    for line in lines:
        if "A*" in line and "(" in line:
            try:
                rounds = int(line.split("(")[1].split(",")[0].strip())
                results.append(rounds)
            except ValueError:
                continue  # Skip invalid lines
    # Replace -2 with "Fail to Solve"
    df = pd.DataFrame(results, columns=["Rounds"]).replace(-2, "Fail to Solve")
    dataframes.append(df)

# Combine all dataframes into one
combined_df = pd.concat(dataframes, axis=1)
combined_df.columns = ["My Results", "Ofek Results", "Sonya Results", "Lior_Results"]  # Update columns

# Add a total row, ignoring "Fail to Solve" entries
def calculate_total(col):
    # Filter out non-numeric entries
    numeric_col = pd.to_numeric(col, errors='coerce')
    return numeric_col.sum(skipna=True)

# Compute the total row for numeric columns
total_row = combined_df.apply(calculate_total).rename("Total")
combined_df.loc["Total"] = total_row

# Adjust index to start from 1
combined_df.index = [i + 1 for i in range(len(combined_df) - 1)] + ["Total"]

# Highlight shortest, middle, and longest rounds
def highlight_rounds(val):
    if val.name != "Total":  # Exclude the Total row
        numeric_vals = [v for v in val if isinstance(v, (int, float))]
        if len(numeric_vals) < 3:  # Skip if there aren't enough numeric values to compare
            return [""] * len(val)
        min_val = min(numeric_vals)
        max_val = max(numeric_vals)
        mid_val = sorted(numeric_vals)[1]
        return ["background-color: green" if v == min_val
                else "background-color: red" if v == max_val
                else "background-color: yellow" if v == mid_val
                else "" for v in val]
    else:
        return [""] * len(val)  # No highlighting for Total row

# Apply styling and display
styled_df = combined_df.style.apply(highlight_rounds, axis=1)
display(styled_df)
```
